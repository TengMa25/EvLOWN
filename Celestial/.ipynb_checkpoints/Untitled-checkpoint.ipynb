```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
jupiter = loadmat('./jupiter_data.mat')['jupiter_data']
saturn = loadmat('./saturn_data.mat')['saturn_data']
```

```python
fig,ax = plt.subplots(figsize = (4,4))
ax.plot(jupiter[0],jupiter[1],lw = .1)
ax.plot(saturn[0],saturn[1],lw = .1)
print("jupiter center: (%.4f, %.4f)"%(np.mean(jupiter[0]), np.mean(jupiter[1])))
print("saturn center: (%.4f, %.4f)"%(np.mean(saturn[0]), np.mean(saturn[1])))
```

```python
jupiter_x = jupiter[0]
jupiter_y = jupiter[1]

jupiter_r = np.sqrt(jupiter_x**2 + jupiter_y**2)
jupiter_rho = 1/jupiter_r
jupiter_theta = np.unwrap(np.arctan2(jupiter_x,jupiter_y))


saturn_x = saturn[0]
saturn_y = saturn[1]

saturn_r = np.sqrt(saturn_x**2 + saturn_y**2)
saturn_rho = 1/saturn_r
saturn_theta = np.unwrap(np.arctan2(saturn_x,saturn_y))
```
