```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

datadir = "../../Data/TJ-3 XHM/Vibration/MEIZHEYVAN/"
case = "B2S11"

def get_vibration(datadir,case):
    data = []
    for i in range(4):
        filename = "%s/%s#%d.txt"%(datadir,case,i+1)
        d = pd.read_csv(filename)
        data.append(d)
    data = np.array(data)
    return np.mean(data,axis = 0)
start = 34#1 
T = 130
Fs = 300

bias = 250

y = get_vibration(datadir,case)[start*Fs+bias:(start+T)*Fs+bias]

y = y-np.mean(y)
print(y[:10])
t = np.linspace(0,len(y)/Fs,len(y))
T = t[-1]
plt.plot(t,y,lw = 0.1)
# plt.xlim(0,10)
# plt.ylim(-2,2)
```

```python
data = {'t':t,'x':y.reshape(-1)}
print(data)
data = pd.DataFrame(data)
data.to_csv('./Data/U307_B2S.csv')
```
