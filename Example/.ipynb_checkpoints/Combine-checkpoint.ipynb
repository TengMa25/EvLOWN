```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from IPython.display import display, Markdown
import sys,os
import numpy as np
from PIL import Image

xsize = 18
ysize = 16

mpl.rc('text', usetex=True)
mpl.rc('font', family='sans-serif')
mpl.rc('xtick', labelsize=4)
mpl.rc('ytick', labelsize=4)
mpl.rc('axes', labelsize=6)
mpl.rc('axes', titlesize=6)
mpl.rc('axes', linewidth=0.5)
mpl.rc('figure', figsize=(xsize/2.54, ysize/2.54))
mpl.rc('figure', dpi=400)
mpl.rc('legend',frameon = False)

project_dir = "Fig/"
sys.path.append(project_dir)
print(os.getcwd())
```

```python
System_names = ["Harmonic","Van der pol","Duffing","Weakly coupled"]
Step_names = ["Weakly\noscillator","Observation","Evolutionary \n library","Evolutionary \n function","Reconstructed \n oscillator"]
figsize = np.array([[3.6,2.8],[3.6,2.8],[3.6,2.8],[3.6,2.8],[3.6,5]])
```

```python
fig,ax = plt.subplots(dpi = 300)
ax.set_xlim(0,xsize)
ax.set_ylim(0,ysize)
ax.set_axis_off()

for i in range(len(Step_names)):
    ax.text(3.6*(i)+2, ysize,Step_names[i],ha="center",va = 'center',fontsize = 7)

for i in range(len(System_names)):
    y_start = ysize-(np.sum(figsize[:i+1,1])+i*0.6)-0.4
    y_end = ysize-(np.sum(figsize[:i,1])+i*0.6)-0.4
    y_middle = (y_start+y_end)/2
    ax.text(0, y_middle,System_names[i],ha="center",va = 'center',fontsize = 7,rotation = 'vertical')

for i in range(4):
    for j in range(5):
        img = Image.open("./Fig/Example%d-%d.jpg"%(i+1,j+1))
        x_start = np.sum(figsize[:j,0])
        x_end = np.sum(figsize[:j+1,0])
        y_start = ysize-(np.sum(figsize[:i+1,1])+i*0.6)-0.4
        y_end = ysize-(np.sum(figsize[:i,1])+i*0.6)-0.4
        ax.imshow(img,extent = (x_start,x_end,y_start,y_end))
    if i != 3:
        ax.plot([0,18],[y_start-0.3,y_start-0.3],color = 'gray',lw = 0.7,ls = '--')

plt.savefig('../../Fig/Figure2. Example.jpg',dpi = 300,transparent = True,bbox_inches="tight",pad_inches = 0.01)
```
